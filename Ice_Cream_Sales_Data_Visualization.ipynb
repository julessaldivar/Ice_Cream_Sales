```python
# Question 3: The ice-cream shop has also aggregated their sales data by month and ice-cream flavor for the previous year
# read df
df = pd.read_csv(r"C:\Users\jules_7owst96\code\Grand Circus\Week 4\Monthly Ice Cream Shop Sales by Flavor.csv")
df.head()
```

```python
# 3A. Create a line graph that displays the total units across all months (time on x axis). Make sure you have the appropriate labels for the x and y axis
plt.plot(df['Month'], df['Total Units'])
plt.title('Total Units of Ice Cream Sold Over the Months')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.grid(True)
plt.show()
```

```python
# Optional: Create another graph with a separate line for each ice cream flavor to compare sales by flavors over time.
plt.plot(df['Month'], df['Vanilla'], label='Vanilla', color='blue')
plt.plot(df['Month'], df['Chocolate'], label='Chocolate', color='brown')
plt.plot(df['Month'], df['Strawberry'], label='Strawberry', color='pink')
plt.plot(df['Month'], df['Neapolitan'], label='Neapolitan', color='green')
plt.plot(df['Month'], df['Mango'], label='Mango', color='orange')
plt.title('Sales of Each Ice Cream Flavor Over the Months')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 3B. Create a bar chart that displays the total profit per month. Make sure you have the appropriate labels for the x and y axis.
sns.barplot(x=df['Month'], y=df['Total Profit'])
plt.title('Total Profit per Month of Ice Cream Sales')
plt.xlabel('Month')
plt.ylabel('Total Profit $')
plt.grid(True)
plt.show()
```
